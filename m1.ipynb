```python
from matplotlib import pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
#sin graph
x=np.linspace(0,np.pi,100)
y=np.sin(x**3)
print(x,y)
```

```python
fig,a = plt.subplots()
a.plot(x,y)
```

```python
#working with subplots
fig, b=plt.subplots(3)
```

```python
x=np.linspace(0,np.pi,100)
y=np.sin(x**3)

fig, b=plt.subplots(3)

b[0].plot(x,y,"tab:orange")
b[1].plot(x,-y,"tab:red")
b[2].plot(-x,y,"tab:green")
```

```python
#neat and subplot_adjust
x=np.linspace(0,np.pi,100)
y=np.sin(x**3)

fig, b=plt.subplots(3)
plt.subplots_adjust(left=0.2,wspace=2,hspace=2,top=0.5,bottom=0.2)

b[0].plot(x,y,"tab:orange")
b[1].plot(x,-y,"tab:red")
b[2].plot(-x,y,"tab:green")
```
